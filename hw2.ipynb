```python
import nltk
```

```python
#1. Грамматика
sentences = ['Вася читает мою книгу', 'Напиши какое-нибудь письмо', 'Этот весёлый мальчик идёт', 'Он любит читать всякие книги']
rules = """
    S -> NP VP
    NP -> ADJ N | N
    VP -> V NP | VP N | V
    ADJ -> 'мою' | 'какое-нибудь' | 'Этот' | 'весёлый' | 'всякие'
    N -> 'Вася' | 'книгу' | 'письмо' | 'мальчик' | 'Он' | 'книги'
    V -> 'читает' | 'Напиши' | 'идёт' | 'любит'| 'читать'
""".split('\n')

grammar = nltk.CFG.fromstring('\n'.join(rules))
cp = nltk.EarleyChartParser(grammar)
```

```python
#2. Проверка 
def print_parses(parser, sentence):
    sentence = sentence.split()
    for tree in parser.parse(sentence):
        print(tree)

for sent in sentences:
    print_parses(cp, sent)
```

```python
cp_verbose = nltk.EarleyChartParser(grammar, trace=1)
for sent in sentences:
    print_parses(cp_verbose, sent)
```
